# 9. 데이터 선택(조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col = '지원번호' )
df
```

```python
df['키'] >= 185
```

```python
df[df['키'] >= 185]
```

```python
filt = df['키'] >= 185
```

```python
df[filt]
```

```python
df[~filt]
```

```python
df.loc[df['키'] >= 185, '수학']
```

### & and

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')]
```

### | 또는

```python
df.loc[(df['키'] >= 190) | (df['키'] <= 185)]
```

###  str 함수

```python
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
filt = df['이름'].str.contains('태')
df[filt]
```

```python
langs = ['PYTHON', 'Java']
filt = df['SW특기'].isin(langs)
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]
```
