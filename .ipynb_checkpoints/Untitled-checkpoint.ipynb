```python
import pandas as pd
import requests
```

```python
code = pd.read_html('https://kind.krx.co.kr/corpgeneral/corpList.do?method=download', header=0)[0]
```

```python
code.sort_values(['상장일'], ascending=True)
```

```python
code = code[['회사명','종목코드']]
code
```

```python
code = code.rename(columns={'회사명':'company', '종목코드': 'code'})
code.head()
```

```python
code.code = code.code.map('{:06d}'.format)
```

```python
code
```

```python
company = '카카오'

companycode = code[code.company==company].code.values[0].strip()

page = 1
```

```python
url = 'https://finance.naver.com/item/sise_day.nhn?code={code}'.format(code=companycode)
url = '{url}&page={page}'.format(url=url, page=page)
print(url)
```

```python
res = requests.get(url,headers= {'User-agent': 'Mozilla/5.0'})
df = pd.read_html(res.text, header=0)[0]
df.head()
```

```python
df = pd.DataFrame()

for page in range(1, 21):
    url = 'https://finance.naver.com/item/sise_day.nhn?code={code}'.format(code=companycode)
    url = '{url}&page={page}'.format(url=url, page=page)
    
    res = requests.get(url,headers= {'User-agent': 'Mozilla/5.0'})
    df = df.append(pd.read_html(res.text, header=0)[0],ignore_index=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df = df.rename(columns={'날짜':'date','종가':'close','전일비':'diff','시가':'open','고가':'high','저가':'low','거래량':'volume'})
```

```python
df[['close', 'diff', 'open', 'high', 'low', 'volume']] = df[['close', 'diff', 'open', 'high', 'low', 'volume']].astype(int)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.sort_values(by=['date'], ascending=True, inplace=True)
df.reset_index(inplace=True)
```

```python
df=df[['date','close','diff','open','high','low','volume']]
```
