# 2. 축

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
```

```python
x = [1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
plt.title('그래프', fontdict={'family' : 'HYGungSo-Bold'})
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x, y)
plt.xlabel('X축', loc='right') # right, center, left
plt.ylabel('Y축', loc='top') # top center bottom
```

```python
plt.plot(x, y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
```
