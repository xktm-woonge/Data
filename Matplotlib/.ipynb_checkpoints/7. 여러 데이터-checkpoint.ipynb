## COVID-19 백신 종류별 접종 인구

```python
days = [1,2,3]
az = [2, 4, 8] # 단위 만명
pfizer = [5,1,3]
moderna = [1,2,3]


plt.plot(days, az)
plt.plot(days, pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='az')
plt.plot(days, pfizer, label='pfizer', marker='o')
plt.plot(days, moderna, label='moderna', marker='s')

plt.legend()
```
