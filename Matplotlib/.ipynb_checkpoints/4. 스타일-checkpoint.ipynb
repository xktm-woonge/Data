# 4. 스타일

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x = [1,2,3]
y = [2, 4, 5]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth = 15)
```

```python
plt.plot(x,y, marker = 'o')
```

```python
plt.plot(x,y, marker = 'o', linestyle = 'None')
```

```python
plt.plot(x,y, marker = 'v', markersize = 10)
```

```python
plt.plot(x,y, marker = 'X', markersize = 10)
```

```python
plt.plot(x,y, marker = 'X', markersize = 10, markeredgecolor='red')
```

```python
plt.plot(x,y, marker = 'o', markersize = 10, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.')
```

## 색깔

```python
plt.plot(x,y,color='pink')
```

## 포맷

```python
plt.plot(x,y ,'ro--')
```

```python
plt.plot(x,y ,'bv--')
```

```python
plt.plot(x,y ,'go')
```

## 축약어

```python
plt.plot(x,y, marker='o', mfc='red', ms=10, mec='blue', ls=':')
```

##  투명도

```python
plt.plot(x,y, marker='o', mfc='red', ms=10,  alpha=0.3)
```

## 그래프 크기

```python
plt.figure(figsize=(10, 5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10, 5), dpi=200)
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
