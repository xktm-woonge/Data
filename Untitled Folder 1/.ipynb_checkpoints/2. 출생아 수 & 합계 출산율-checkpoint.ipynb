# 2. 출생아 수 & 합계 출산율

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('stat_142801.xls', skiprows=2, nrows=2, index_col = 0)
```

```python
df
```

```python
df.index.values
```

```python
df.rename(index={'출생아\xa0수': '출생아 수', '합계\xa0출산율' : '합계 출산율'}, inplace=True)
```

```python
df = df.T
```

```python
df
```

```python
plt.plot(df.index, df['출생아 수'])
```

```python
plt.plot(df.index, df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(10,10))
ax1.plot(df.index, df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx()
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize=(10 , 5))
ax1.set_ylabel('출생아 수(천명)')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')

ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize=(10 , 5))
ax1.set_ylabel('출생아 수(천명)')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 15, val, ha='center')

ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize=(10 , 5))
ax1.set_ylabel('출생아 수(천명)')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 15, val, ha='center')

ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o', ms=10, lw=3, mec='w', mew=1)
```

```python
fig, ax1 = plt.subplots(figsize=(10 , 5))
fig.suptitle('출생아 수 및 합계 출산율')
ax1.set_ylabel('출생아 수(천명)')
ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax1.bar(df.index, df['출생아 수'], color='#ff812d')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 15, val, ha='center')

ax2 = ax1.twinx()
ax2.set_ylabel('합계 출산율')
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0,1])
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o', ms=10, lw=3, mec='w', mew=1)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.07, val, ha='center')
```
