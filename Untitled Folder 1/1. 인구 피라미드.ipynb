```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

## 2012남자 데이터

```python
df_m = pd.read_excel('201202_201202_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,E:Y')
```

```python
df_m
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int)
```

```python
df_m
```

## 2012여자 데이터

```python
df_w = pd.read_excel('201202_201202_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,AB:AV')
```

```python
df_w.columns = df_m.columns
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
```

## 시각화

```python
plt.figure(figsize=(10,10))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000)
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2012 인구 피라미드')
plt.show()
```
